```python
#### FUNCTIONS ######
def createGraph(row):
    sender = row.SenderFirstName +"."+ row.SenderLastName
    receiver = row.ReceiverFirstName +"."+ row.ReceiverLastName
    
    # No self edges
#     if sender == receiver: 
#         return None
    # Add nodes for the sender and receiver if they don't already exist
    if not DG.has_node(sender):
        DG.add_node(sender,{"EID":row.SenderEID,"Email":row.SenderEmail,"Position":str(row.Position_s),"Role":str(row.Role_s)})

    if not DG.has_node(receiver):
        DG.add_node(receiver,{"EID":row.ReceiverEID,"Email":row.ReceiverEmail,"Position":str(row.Position_r),"Role":str(row.Role_r)})

    # Add edge if it doesn't exist
    if not DG.has_edge(sender,receiver):
        DG.add_edge(sender,receiver,{"MID":row.mid,"SentDate":row.SentDate,"Type":row.ReceiverType})

    return None

def rightEigenVector(A,x0,max_iters=2000):
    """
    Power Method to find the leading eigenvector
    """
    xt = x0.copy()
    i = 0
    while True:
        x = xt
        xt = np.dot(A,x)
        i = i + 1

        # renormalize
        x = x / np.linalg.norm(x)
        xt = xt / np.linalg.norm(xt)

        if np.allclose(x,xt) or  i > max_iters:
            break
    
    return xt

def leftEigenVector(A,x0,max_iters=2000):
    xt = x0.copy()
    i = 0
    while True:
        x = xt
        xt = np.dot(x.T,A)
        i = i + 1

        # renormalize
        x = x / np.linalg.norm(x)
        xt = xt / np.linalg.norm(xt)

        if np.allclose(x,xt) or  i > max_iters:
            break
    
    return xt

def plotCentrality(centrality,max_names_show=10,title=None,filename=None):
    sortedCenNames = sorted(centrality.keys(),key=lambda x: centrality[x])[-max_names_show:]
    sortedCenValues = [centrality[s] for s in sortedCenNames][-max_names_show:]

    # plot it
    fig = plt.figure(figsize(10,7))
    ax = plt.subplot(111)
    plt.barh(np.arange(len(sortedCenValues)),sortedCenValues,alpha=0.5)
    ax.set_yticks(0.5 + np.arange(len(sortedCenValues)))
    ax.set_yticklabels(sortedCenNames)
    if title is not None: plt.title(title)
    if filename is not None: plt.savefig(filename,format="png")

    
def getCutsize(G,n1,n2):
    count = 0
    for i in n1:
        for j in n2: 
            if G.has_edge(i,j): count += 1
    return count

def getModularityMatrix(G):
    k = nx.degree(G)
    nodes = G.nodes()
    m = len(nx.edges(G))
    
    arrayK = np.array([float(k[n]) for n in nodes])
    A = np.array(nx.to_numpy_matrix(G))
    K = np.outer(arrayK,arrayK)/(2*m)
    return (A - K)

def getGenModularityMatrix(G,nG):
    B = getModularityMatrix(G)
    nodes = G.nodes()
    
    # create empty matrix
    BG = np.zeros((len(nG),len(nG)))
    
    for i in range(len(nG)):
        rowSum = 0
        for j in range(len(nG)):
            a = nodes.index(nG[i])
            b = nodes.index(nG[j])
            BG[i,j] = B[a,b]
            rowSum += B[a,b]
        BG[i,i] = BG[i,i] - rowSum
    return BG
        
def partitionSubGraph(G,nX):
    BGX = getGenModularityMatrix(G,nX)
    xX = rightEigenVector(BGX,np.ones(BGX.shape[1]))
    nX1 = [nX[i] for i in range(len(nX)) if xX[i] > 0]
    nX2 = [nX[i] for i in range(len(nX)) if xX[i] <= 0]
    
    if len(nX1) < 2 or len(nX2) < 2 :
        return {random.random():nX1,random.random():nX2}
    else:
        return dict(partitionSubGraph(G,nX1).items() + partitionSubGraph(G,nX2).items())
    
def partitionGraph(G):
    B = getModularityMatrix(G)
    nodes = G.nodes()

    x = rightEigenVector(B,np.ones(B.shape[1]))
    n1 = [nodes[i] for i in range(len(nodes)) if x[i] > 0]
    n2 = [nodes[i] for i in range(len(nodes)) if x[i] <= 0]
    return dict(partitionSubGraph(G,n1).items() + partitionSubGraph(G,n2).items())

def plotGraphColorNodes(G,nListOfList,figsize=(20,20),alpha=0.8,nodesize=100,font_size=10):
    pos = nx.spring_layout(G)
    nodes = G.nodes()
    colors = np.linspace(0,1,len(nListOfList))
    
    c = np.zeros(nx.number_of_nodes(G))
    for nList,i in zip(nListOfList,range(len(nListOfList))):
        for n in nList: 
            c[nodes.index(n)] = colors[i]        
    
    plt.set_cmap("Accent")
    fig = plt.figure(figsize=figsize)
    nx.draw_networkx(G,nodesize=nodesize,font_size=font_size,node_color=c,alpha=alpha)
```

```python

###### Spectral Partitioning #####
# 1. Calculate the eigenvector v2 corresponding to the second smallest eigenvalue λ2 of the
# graph Laplacian.
# 2. Sort the elements of the eigenvector in order from largest to smallest.
# 3. Put the vertices corresponding to the n1 largest elements in group 1, the rest in group 2, and
# calculate the cut size.
# 4. Then put the vertices corresponding to the n1 smallest elements in group 1, the rest in group
# 2, and recalculate the cut size.
# 5. Between these two divisions of the network, choose the one that gives the smaller cut size.

# get the laplacian of the undirected matrix
G = DG.to_undirected()
sparseL = nx.linalg.laplacian_matrix(G)
print type(sparseL)
L = nx.linalg.laplacian_matrix(G).toarray()
nodes = G.nodes()
n1_size = int(len(nodes)*0.5)

# get the eigenvalues
eigVal,eigVec = np.linalg.eig(L)
orderEigVal = np.argsort(eigVal)

# find the second lowest eigenvalue that is not 0
print eigVal[orderEigVal[0]], eigVal[orderEigVal[1]], eigVal[orderEigVal[2]],eigVal[orderEigVal[3]]
v = eigVec[:,orderEigVal[2]]
plt.figure(figsize=(9,9))
plt.scatter(range(len(eigVal)),eigVal)
plt.title("Eigenvalues of the Laplacian Matrix")

# spilt according to the magnitude of eigenvalue
orderV = np.argsort(v)
n1 = [nodes[n] for n in orderV[:n1_size]]
n2 = [nodes[n] for n in orderV[n1_size:]]


# highlight important nodes
pos = nx.spring_layout(G)
c = ["blue"]*nx.number_of_nodes(G)
for n in n1: c[nodes.index(n)] = "red"

fig = plt.figure(figsize=(20,20))
nx.draw_networkx(G,nodesize=100,font_size=10,node_color=c,alpha=0.25)
plt.title("Graph Split into 2 Groups")
```
