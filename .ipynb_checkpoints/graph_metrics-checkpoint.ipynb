```python
import numpy as np
import pandas as pd
import networkx as nx
import datetime as dt
import matplotlib.pyplot as plt
```

```python
DG = nx.DiGraph()

def createGraph(row):
    sender = row.SenderFirstName +"."+ row.SenderLastName
    receiver = row.ReceiverFirstName +"."+ row.ReceiverLastName
    
    # Add nodes for the sender and receiver if they don't already exist
    if not DG.has_node(sender):
        DG.add_node(sender,{"EID":row.SenderEID,"Email":row.SenderEmail})

    if not DG.has_node(receiver):
        DG.add_node(receiver,{"EID":row.ReceiverEID,"Email":row.ReceiverEmail})

    # Add edge if it doesn't exist
    if not DG.has_edge(sender,receiver):
        DG.add_edge(sender,receiver,{"MID":row.mid,"SentDate":row.SentDate,"Type":row.ReceiverType})

    return None
```

```python
# get the datafile
data = pd.read_csv("enronDataCleaned.csv")
data["SentDate"] = data.SentDate.apply(lambda x: dt.datetime.strptime(x,"%Y-%m-%d %H:%M:%S"))
data["Year"] = data.SentDate.apply(lambda x: x.year)
data["Month"] = data.SentDate.apply(lambda x: x.month)
data["Day"] = data.SentDate.apply(lambda x: x.day)

#subset and create graph
data2 = data[(data["Year"] == 2000) & (data["Month"] == 10)]
success = data2.apply(createGraph,1)
```

```python
fig = plt.figure(figsize=(20,20))
nx.draw_networkx(DG,nodesize=100,font_size=10,node_color="blue",alpha=0.25)
```

```python
# Get Eigenvalue centrality
eigCentrality = nx.eigenvector_centrality_numpy(DG)
sortedEigCenNames = sorted(eigCentrality.keys(),key=lambda x: eigCentrality[x])
sortedEigCenValues = [eigCentrality[s] for s in sortedEigCenNames]

# plot it
fig = plt.figure(figsize=(16,16))
ax = plt.subplot(111)
a =  plt.barh(np.arange(len(sortedEigCenValues)),sortedEigCenValues)
a = ax.set_yticks(np.arange(len(sortedEigCenValues)))
a = ax.set_yticklabels(sortedEigCenNames)

```

```python
nodes = DG.nodes()
eigCenComputed = {}

A = nx.to_numpy_matrix(DG)
eigVal,eigVec = np.linalg.eig(A)

# find the maximum Eigenvalue
orderedEigVal = np.argsort(eigVal)
kappaMax = eigVec[:,orderedEigVal[-1]]
sumK = sum(kappaMax)

for i,name in zip(range(len(nodes)),nodes):
    eigCenComputed[name] = kappaMax[i,0] /sumK

sortedEigCenNames = sorted(eigCenComputed.keys(),key=lambda x: eigCenComputed[x])
sortedEigCenValues = [eigCenComputed[s] for s in sortedEigCenNames]

# plot it
fig = plt.figure(figsize=(16,16))
ax = plt.subplot(111)
a =  plt.barh(np.arange(len(sortedEigCenValues)),sortedEigCenValues)
a = ax.set_yticks(np.arange(len(sortedEigCenValues)))
a = ax.set_yticklabels(sortedEigCenNames)
```
